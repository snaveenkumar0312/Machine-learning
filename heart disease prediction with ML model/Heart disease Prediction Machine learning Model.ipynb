```python
#separate the target fron dataframe
x=heart_data.drop('target',axis=1)
y=heart_data.target
```

```python
## standardize the data
import numpy as np
from sklearn.preprocessing import StandardScaler
StdScaler_model=StandardScaler()
StdScaler_model.fit_transform(x)
```

```python
#spliting the data for testing & training
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=2)
print(X_train.shape , X_test.shape, y_train.shape, y_test.shape)
```

```python
##import the model from sklearn library
from sklearn.linear_model import LogisticRegression 
logistic_model=LogisticRegression()
logistic_model.fit(X_train, y_train)
```

```python
##model has learn and pridect for test data
predicted_values=logistic_model.predict(X_test)
predicted_values
```

```python
real_values=y_test
real_values
```

```python
## now the accuracy score will find ths difference between the "predicted_values" & "real_values"
from sklearn.metrics import accuracy_score
accuracy_model=accuracy_score(predicted_values,real_values)
accuracy_model
```
